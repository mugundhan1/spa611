```python
import numpy as np
import matplotlib.pyplot as plt
```

Q1) Verify that the mean of power is centered around the variance of the voltage

For a zero mean random voltage signal V, power is defined as P = V².

The mean power is:
E[P] = E[V²]
Variance is:
Var(V) = E[V²] - (E[V])²

For zero mean signals, E[V] = 0, therefore:

Mean Power = Variance

Now, we will verify it:

```python
N = 10000
sigma = 2

V = np.random.normal(0, sigma, N)
P = V**2

mean_power = np.mean(P)
variance_voltage = np.var(V)

print("Mean Power =", mean_power)
print("Variance =", variance_voltage)

plt.figure(figsize=(6,4))
plt.hist(P, bins=50)
plt.title("Histogram of Power")
plt.xlabel("Power")
plt.ylabel("Frequency")
plt.show()
```

It is verified that mean power and variance are equal

Q2) Consider cases of different RMS and plot their histograms

As RMS increases, the histogram becomes wider. This indicates increased signal spread and higher variance.

```python
sigma_values = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(6,4))

for sigma in sigma_values:
    V = np.random.normal(0, sigma, N)
    plt.hist(V, bins=50, alpha=0.5, label=f'RMS={sigma}')

plt.legend()
plt.title("Histogram for different RMS values")
plt.xlabel("Voltage")
plt.ylabel("Frequency")
plt.show()
```

Q3) How will the histogram change if I add a slowly varying sine wave in it

```python
t = np.linspace(0, 10, N)
noise = np.random.normal(0, 2, N)
sine = 2*np.sin(0.5*t)

V_new = noise + sine

plt.figure(figsize=(6,4))
plt.hist(V_new, bins=50)
plt.title("Histogram with sine wave added")
plt.show()
```

The sine wave shifts voltage values, causing the histogram to widen and change shape.

Q4) How are the Variances of the ensembles distributed ?

```python
variances = []

for i in range(100):
    V = np.random.normal(0, 2, N)
    variances.append(np.var(V))

plt.figure(figsize=(6,4))
plt.hist(variances, bins=30)
plt.title("Distribution of Ensemble Variances")
plt.show()
```

The variances are distributed around the true variance value due to statistical fluctuations. The simulation confirms theoretical relationships between power, RMS, and variance. 
Mean power equals variance for zero mean signals. Increasing RMS increases signal spread. 
Adding sine wave alters distribution, and ensemble variances cluster around true value.
Through these graphs, we could clearly see the theory coming alive — the mean power matched the variance for zero-mean noise, and squaring Gaussian noise gave the expected skewed power pattern. When we added a slow sine wave, the signal energy increased and the shape changed, but repeated trials still showed the variance estimates staying close to the true value, just with small natural fluctuations.
